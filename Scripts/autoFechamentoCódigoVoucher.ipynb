```python
# Importando bibliotecas.
import pandas as pd
import pyautogui as pygui
import pyperclip
import time
import re
```

```python
# Abrindo arquivo xlsx.
arquivo = pd.read_excel("fechamento1.xlsx")
arquivoPedidoExt = arquivo["pedidoExt"]
```

```python
# Regex para unidade.
unidadeString = re.compile(r'UNIDADE')
# Regex para analitica.
analiticaString = re.compile(r'ANALÍTICA')
# Verificação PENDENTE.
pendenteString = "PENDENTE"
```

```python
# Função Ar Connect (Use na página 90%)
def autoArConnect(codigo):
    
    time.sleep(2)

    # Clica no botão filtro.
    pygui.click(x=430, y=193, button="left")
    time.sleep(1)

    # Clica no TextField.
    pygui.click(x=489, y=420, button="left")
    time.sleep(1)

    # Seleciona todo o texto.
    pygui.hotkey("ctrl", "a")
    time.sleep(0.5)  # Espera um pouco para garantir a seleção

    # Escreve o código.
    pygui.write(str(codigo))  # Garante que o código seja uma string
    time.sleep(0.5)  # Espera um pouco após escrever

    # Clica em pesquisar.
    pygui.click(x=883, y=543, button="left")
    time.sleep(4)  # Espera a página carregar

    # Seleciona editar informação.
    pygui.click(x=1249, y=243, button="left")
    time.sleep(5)

    # Clica no campo voucher.
    pygui.click(x=106, y=332, button="left")
    # Seleciona o texto.
    pygui.hotkey("ctrl", "a")
    time.sleep(0.5)
    # Copia o texto.
    pygui.hotkey("ctrl", "c")
    time.sleep(0.5)

    pendente = pyperclip.paste()
    
    # Verifica se há o status pendente.
    if pendente == pendenteString:

        pygui.press("down")
        pygui.press("down")
        pygui.press("down")
        pygui.press("down")

        # Seleciona o texto da "Unidade".
        pygui.click(x=95, y=708, button="left")
        time.sleep(0.5)
        # Seleciona o texto.
        pygui.hotkey("ctrl", "a")
        time.sleep(0.5)
        # Copia o texto.
        pygui.hotkey("ctrl", "c")
        time.sleep(0.5)

        pygui.scroll(500)
        time.sleep(0.5)
        
    else: 
        # Seleciona o texto da "Unidade".
        pygui.click(x=158, y=701, button="left")
        time.sleep(0.5)
        # Seleciona o texto.
        pygui.hotkey("ctrl", "a")
        time.sleep(0.5)
        # Copia o texto.
        pygui.hotkey("ctrl", "c")
        time.sleep(0.5)

    # Valor da unidade.
    unidade = pyperclip.paste()

    # Clica em voltar.
    pygui.click(x=1314, y=148, button="left")
    time.sleep(3)
    print(unidade)

    return str(unidade)

autoArConnect("72698144")
```

```python
# Substituindo/Adicionando parceiro.
for index, codigo in enumerate(arquivoPedidoExt):
    
    if index != 5:
        try:
            # Atribui o valor da unidade encontrado na automação.
            parceiro = autoArConnect(codigo)
            parceiro = re.sub(unidadeString, '', parceiro).strip()
            print(f"parceiro: {parceiro}, index: {index}")
            # Atribui o valor em maiúsculo ao indice da planilha.
            arquivo.at[index, "parceiro"] = parceiro.upper()
        except Exception as e:
            print(f"Erro ao salvar o arquivo {e}, no indice: {index}")

arquivo
arquivo.to_excel("fechamento1Completo.xlsx")
```
